```python
# Code by : Chhayan Thanki (216270307017)

# Problem statment - Write a Pandas program to create a line plot of the opening, closing stock prices of given company between two specific dates.

# Importing needed libreries 
import pandas as pandas # To perform pandas series operations
import matplotlib.pyplot as pyplot # To plot analysis charts 
```

```python
# Reading Apple stock dataset csv file
data_set = pandas.read_csv("../input/apple-stock-data-live-and-latest-from-ipo-date/Apple_stock_history.csv")
```

```python
# Specify starting date and ending date for line chart
start_date = pandas.to_datetime('2020-4-1')
end_date = pandas.to_datetime('2020-09-30')

# Convert string Date time into Python Date time object.
data_set['Date'] = pandas.to_datetime(data_set['Date']) 

# Create new data frame with records between starting date and ending date
new_df = (data_set['Date']>= start_date) & (data_set['Date']<= end_date)
df2 = data_set.loc[new_df]
```

```python
# Ploting the line plot
df2.plot(x='Date', y=['Open', 'Close']);
pyplot.suptitle('01-04-2020 to 30-09-2020', fontsize=12, color='black')

#  Labels for line plot 
pyplot.xlabel("Date",fontsize=12, color='black')
pyplot.ylabel("Price", fontsize=12, color='black')
pyplot.show()
```
